```python
import sys
import numpy as np
import matplotlib.pyplot as plt
sys.path.insert(0, "/Users/lschumacher/Documents/GitHub/superstats")

from superstats.transition.rw import RW
from superstats.prior.prior import Prior
from superstats.prior.joint_prior import JointPrior
```

```python
prior = JointPrior(
    v=RW(bounds=(0, 3), initial_prior=Prior("normal", loc=0, scale=1)),
    a=RW(bounds=(0, 3), initial_prior=Prior("normal", loc=0.2, scale=1)),
    ndt=Prior("uniform", low=0, high=0.5)                             
)
```

```python
prior_samples = prior.sample(steps=5)
```
